```python
import pandas as pd

# Read the file as raw text
raw = pd.read_csv('/content/preventativescan_member_feedback.csv', header=None, dtype=str)

# Remove the outer quotes from each line
raw[0] = raw[0].str.strip('"')

# Replace doubled quotes ("") with single quotes (")
raw[0] = raw[0].str.replace('""', '"')

# Now split each line into columns
df = raw[0].str.split(',', n=4, expand=True)

# Assign column names
df.columns = ['response_id', 'nps_score', 'feedback_text', 'response_date', 'member_location']

df.head()
df.columns
```

```python
df['feedback_text'].to_csv('feedback_only.csv', index=False)
```
